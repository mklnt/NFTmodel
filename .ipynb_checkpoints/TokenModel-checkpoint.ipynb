```python
#
#R=1/10 #percentage of investors each year 

#time which we want to analyze
Y=6. # number of years
dY=12. # intervals of years (supposing equal onboarding probability throughout the year)



#R=R/dY


#N0=5000 #total number of possible investors
T=10000 #number of tokens per investors


#stuff needed for the program to run
S=int(Y*dY) # number of steps
Years=[0]*S
N=[0]*S
#N[0]=N0
Inv=[0]*S #new investors
InvTot=[0]*(S+2)  #total investors
#print(N)
#print(Inv)
#print(S)
Revenue=0
for t in range(0,S):
    Years[t]=t/dY
```

```python
def model(ind):
    N[0]=ind[0]
    Inv[0]=0
    R=ind[1]/dY
    for t in range(1,S):
        Inv[t]=N[t-1]*R
        InvTot[t]=(InvTot[t-1] + Inv[t])  
        N[t]=(N[t-1] - Inv[t]) 
        
#When do we reach 400 investors?        
        
        if InvTot[t] <= 400:
            InvTot[S]=InvTot[t]
            InvTot[S+1]=t/dY
           
    return InvTot #number of total investors per period of time
```

```python
def show():
    fig=plot(InvTot[S+1],InvTot[S], marker="o", markersize=5 )
    #fig.update_traces(hovertemplate = 'hover_data_0=%{InvTot*T}')
    return fig 
    #,plot.update_traces(hovertemplate = 'hover_data_0=%{InvTot*T}')#, plt.axvline(x=InvTot[S+1], ymin=0, ymax=InvTot[S])
```

```python
#what if the probability of investors is not constant 

def modelNoise(N0,R):
    N[0]=N0
    Inv[0]=0
    R=R/dY
    for t in range(1,S):
        R=R+(randn()/500)
        #print(R)
        Inv[t]=N[t-1]*R
        InvTot[t]=(InvTot[t-1] + Inv[t])  
        N[t]=(N[t-1] - Inv[t]) 
    return InvTot  #number of total investors per period of time
```

```python
figure(figsize=(20,10))
plot(Years, modelNoise(5000,1/10), label="juice industry")
#plot(Years, model(10000, 1/20), label="soya industry")
legend(frameon=False)
xlabel("time")
ylabel("Investors")
tight_layout()
```
